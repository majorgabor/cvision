```python
import numpy as np
from scipy import ndimage
import matplotlib.pyplot as plt
```

```python
img = 1.0 * plt.imread("./filters/einstein.jpg")
```

```python
filter = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
```

```python
filter = np.array([[-0.125, -0.25, -0.125], [0, 0, 0], [0.125, 0.25, 0.125]])
```

```python
img_x = ndimage.convolve(img, filter, mode="reflect")
img_y = ndimage.convolve(img, filter.T, mode="reflect")
img_g = np.sqrt(img_x ** 2 + img_y ** 2)
```

```python
plt.figure(figsize=(20,7))
plt.subplot(1, 3, 1)
plt.imshow(img_x, cmap="gray")
plt.subplot(1, 3, 2)
plt.imshow(img_y, cmap="gray")
plt.subplot(1, 3, 3)
plt.imshow(img_g, cmap="gray")
```

```python
diff = img - img_g
plt.figure(figsize=(20,7))
plt.subplot(1,2,1)
plt.imshow(img, cmap="gray")
plt.subplot(1,2,2)
plt.imshow(diff, cmap="gray")
```
