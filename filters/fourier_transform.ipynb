```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = plt.imread("./einstein.jpg")
F = np.fft.fftshift(np.fft.fft2(img))
F_mag = np.abs(F)
```

```python
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.imshow(F_mag, cmap="gray")
plt.subplot(1,2,2)
plt.imshow(np.log(F_mag), cmap="gray")
```
